```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
### Read the DataSet

df = pd.read_csv("placement.csv")
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])  ## Now, I'm cheking my datasets is linear or sort of linear, so we can see our datasets
plt.xlabel('CGPA')                         ## is completely sort of linear.
plt.ylabel('Package in (LPA)')
```

```python
## Here I'm seprating Independent variable and dependent variable

X = df.iloc[:,0:1]  # Independent variable
y = df.iloc[:,-1]   # dependent varivable
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Spiliting the dataset into training and testing datasets

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# # instantiate the Linear Regression model

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train) # Training the model 
```

```python
print('Accuracy :',lr.score(X_test,y_test))
```

```python
## Here, I am trying to show Best fit line 

plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Package in (LPA)')
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
comparison_df = pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
comparison_df
```
